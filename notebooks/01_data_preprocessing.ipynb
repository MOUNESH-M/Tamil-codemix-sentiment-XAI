```python
import pandas as pd
import unicodedata
import re
df=pd.read_csv("tamil_sentiment_full.csv",sep="\t", on_bad_lines='skip', names=["sentiment","text"])
print("----------------------------------------------------")
print("****DATA BEFORE CLEANING******")
print("----------------------------------------------------")
print(df.head(),"\n")
print(df.sentiment.value_counts())
print("total samples", len(df))

#-------------------------cleaning---------------------------------------
def clean_sentiment_classes(data):
  data["sentiment"] = data["sentiment"].apply(lambda x: unicodedata.normalize("NFKC", str(x)))

  # 2) Strip spaces before/after
  data["sentiment"] = data["sentiment"].str.strip()

  # 3) Replace accidental duplicates with correct spelling
  data["sentiment"] = data["sentiment"].replace({
    "Positive ": "Positive",
    " positive": "Positive",
    "Positve": "Positive",
  })

  # 4) Define allowed labels only
  valid_labels = ["Positive", "Negative", "Mixed_feelings"]

  # 5) Remove rows with invalid labels
  data = data[data["sentiment"].isin(valid_labels)]
  return data

def clean_text(txt):
    txt = str(txt)
    txt = re.sub(r"http\S+", "", txt)
    txt = re.sub(r"@\w+", "", txt)
    txt = re.sub(r"#[A-Za-z0-9_]+", "", txt)
    txt = re.sub(r"[^a-zA-Zஅ-ஹ\s]", "", txt)
    txt = re.sub(r"[^A-Za-z\u0B80-\u0BFF ]+", " ", txt)
    txt = txt.lower().strip()
    return txt

df["text"]=df["text"].apply(clean_text)
df= clean_sentiment_classes(df)
print("----------------------------------------------------")
print("****DATA AFTER CLEANING******")
print("----------------------------------------------------")
print(df.head(),"\n")
print(df["sentiment"].value_counts())
print("total samples", len(df))

#df.to_csv("tamil_sentiment_full.csv")
```
